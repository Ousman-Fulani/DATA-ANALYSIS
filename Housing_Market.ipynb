```python
import pandas as pd
import numpy as np
```

```python
df1=pd.Series(np.random.randint(1,4,size=100))
df2=pd.Series(np.random.randint(1,3,size=100))
df3=pd.Series(np.random.randint(10000,30000,size=100))
```

```python
df=pd.concat([df1,df2,df3,],axis=1)
df
```

```python
df.rename(columns={0:'large',1:'larger',2:'largest'},inplace=True)
df.head()
```

```python
# join concat the values
bigcolumn = pd.concat([df1, df2, df3], axis=0)

# it is still a Series, so we need to transform it to a DataFrame
bigcolumn = bigcolumn.to_frame()
print(type(bigcolumn))

bigcolumn
```

```python
bigcolumn.reset_index(drop=True, inplace=True)
bigcolumn
```
